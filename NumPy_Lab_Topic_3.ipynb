## 🔹 Activity 7: Broadcasting

```python
x = np.array([[1,2,3],[4,5,6]])
v = np.array([1,0,1])

# Predict output before running
print(x + v)
```

```python
import numpy as np

x = np.array([[1,2,3],[4,5,6]])  # Shape: (2, 3)  ← 2 rows, 3 columns
v = np.array([1,0,1])            # Shape: (3,)    ← 1D array (length 3)

print(x + v)  # ❌ BROADCASTING FAILURE

Why it fails - Broadcasting Rules:

x.shape: (2, 3)    v.shape: (3,)
[2 rows][3 cols]   [length 3]

NumPy tries to stretch v.shape to match:
(3,) → (1, 3) → (2, 3)  ?  ❌ Can't stretch 1→2 automatically
```

```python
x = np.array([[1,2,3],[4,5,6]])
v = np.array([1,0,1])

print(x + v)
```

## 🔹 Activity 8: Mini Project

```python
marks = np.array([
    [70, 80, 90],
    [60, 75, 85],
    [50, 65, 70]
])

# TODO:
# 1. Average per student
# 2. Average per subject
# 3. Highest mark
# 4. Students with average > 75
```

```python
import numpy as np

marks = np.array([
    [70, 80, 90],
    [60, 75, 85],
    [50, 65, 70]
])

print("Marks:")
print(marks)

avg_per_student = np.mean(marks, axis=1)
print("\n1. Average per student:", avg_per_student)

avg_per_subject = np.mean(marks, axis=0)
print("2. Average per subject:", avg_per_subject)

highest = np.max(marks)
print("3. Highest mark:", highest)

top_students = np.where(avg_per_student > 75)[0]
print("4. Students with avg > 75 (indices):", top_students)
print("Their averages:", avg_per_student[top_students])
```
